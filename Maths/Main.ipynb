```python
from sympy.vector import *
from sympy import symbols
from pprint import pprint

N = CoordSys3D('N')
x,y,z = symbols('x y z')
A = N.x**2*N.y**2+N.x*N.z-4
delop = Del()
display(delop(A))
gradA = gradient(A)
pprint(f'The gradient of {A} is ')
display(gradA)
```

```python
from sympy import * 
t , s , tau = symbols ('t,s,tau ') 
f = t 
g = exp( t ) 
fog = integrate ( f . subs ({t:tau})*g . subs ({t:t-tau}) , ( tau ,0 , t ) ) 
def LT ( func ): 
    return laplace_transform ( func ,t ,s , noconds = True ) 
FOG = LT ( fog ) 
print ('Laplace transform of the convolution of given functions is:') 
display ( FOG )
```

```python
from sympy import Symbol,lambdify
x= Symbol('x') 
g = (x**3)-(2*x)-5
f = lambdify(x,g) 
a = float (input ('Enter a value:')) 
b = float (input ('Enter b value:')) 
n = int (input ('Enter number of iterations:')) 
for i in range (1, n+1): 
    c = (a*f(b)-b*f(a))/(f(b)-f(a)) 
    if ((f(a) * f(c) < 0)): 
        b=c 
    else: 
        a=c 
    print ('Iteration %d \t the root %0.3f \t function value %0.3f \n'%(i,c,f(c)))
```
