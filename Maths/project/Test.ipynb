```python
from sympy import *
from sympy.vector import CoordSys3D
import Mat
```

## Gradient

```python
N = CoordSys3D('N')
x,y,z = symbols('x y z')
A = N.x**2*N.y**2+N.x*N.z-4
Mat.gradient(A)
```

```python
B = N.x**3*N.y**x+N.x*N.z
Mat.gradient(B)
```

```python
A = N.x**2*N.y*N.z+N.y**2*N.z*N.x+N.z**2*N.y*N.x
Mat.gradient(A)
```

```python
A = N.x**y*N.y*N.z+N.y**3*N.z*N.x+N.z**2*N.y*N.x
Mat.gradient(A)
```

## Curl

```python
N = CoordSys3D('N')
x,y,z = symbols('x y z')
A = N.x**2*N.y*N.z*N.i+N.y**2*N.z*N.x*N.j+N.z**2*N.y*N.x*N.k
Mat.Cur_l(A)
```

```python
A = N.x**y*N.y*N.z*N.i+N.y**y*N.z*N.x*N.j+N.z**2*N.y*N.x*N.k
Mat.Cur_l(A)
```

```python
A = N.x**3*N.y*N.z*N.i+N.y**x*N.z*N.x*N.j+N.z**x*N.y*N.x*N.k
Mat.Cur_l(A)
```

## Divergence

```python
N = CoordSys3D('N')
x,y,z = symbols('x y z')
A = N.x**2*N.y*N.z*N.i+N.y**2*N.z*N.x*N.j+N.z**2*N.y*N.x*N.k
Mat.diver(A)
```

```python
A = N.x**y*N.y*N.z*N.i+N.y**y*N.z*N.x*N.j+N.z**2*N.y*N.x*N.k
Mat.diver(A)
```

## Laplace Transform

```python
a,s,t = symbols('a,s,t')
A = sin(a*t)
B = 1/(s**4)
C = sinh(a*t)
Mat.Laplace_Transforming(A)
print('\n')
Mat.Laplace_Transforming(B)
print('\n')
Mat.Laplace_Transforming(C)
```

## Inverse Laplace Transform

```python
Mat.inverse_Laplace_Transforming(B)
```

## Regula Falsi

```python
x = Symbol('x')
d = (x**3)-(2*x)-5
Mat.RegulA_fAlsi(d, x, 2, 3, 5)
```

## Sin Tan relation

```python
import numpy as np
import matplotlib . pyplot as plt 
ax = plt.figure().add_subplot(projection='3d')
x = np.linspace(0, 1, 100)
y = np.sin(x * 2 * np.pi) / 2 + 0.5
ax.plot(x, y, zs=0, zdir='z', label='Sin in (x, y)')

x = np.linspace(0, 1, 100)
y = np.tan(x * 2 * np.pi) / 2 + 0.5
ax.plot(x, y, zs=0, zdir='y', label='Tan in (x, z)')
plt.legend()
plt.show()
```

## Sin Cos relation
